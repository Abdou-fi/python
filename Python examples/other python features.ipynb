```python
def gcd(a, b): 
    # Function to calculate GCD of two numbers using recursion
    if a < 0 or b < 0:
        print("Please enter non-negative integers.")
        return None
    elif b == 0:
        return a
    else:
        return gcd(b, a % b)
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

result = gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is {result}")
```

Write a Python Program to Find the Factorial of a Number.

```python
# Write a Python Program to Find the Factorial of a Number.
def factorial(num) :
    factorial = 1 
    if type(num) != int or num < 0 :
        return None 
    elif num == 0:
        return 1
    else:
        for i in range(1, num + 1):
            factorial = factorial * i 
        return factorial

num = int(input("Enter a number to find its factorial: "))
print(f"The factorial of {num} is {factorial(num)}")
```

```python
# Write a Python Program to Print all Prime Numbers in an Interval a - b.
lower = 100
upper = 200
print(f"Prime numbers between {lower} and {upper} are:")
for num in range(lower, upper + 1):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```

```python
# fibonacci sequence
def fibonacci(n):
    a, b = 0, 1
    sequence = []
    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence
n = int(input("Enter the number of terms in the Fibonacci sequence: "))
print(f"Fibonacci sequence with {n} terms: {fibonacci(n)}")
```

```python
# Basic number matching
num = int(input("Enter a number: "))
match num:
    case 1:
        print("Number is 1")
    case 2:
        print("Number is 2")
    case 3:
        print("Number is 3")
    case _:
        print("other Number")
        
```

```python
# Matching string commands
command = input("Enter a command (start, stop, pause): ")
match command:
    case "start":
        print("Starting the process...")
    case "stop":
        print("Stopping the process...")
    case "pause":
        print("Pausing the process...")
    case _:
        print("Unknown command")
```

```python
# Tuple matching
point = (10, 20)
match point:
    case (0, 0):
        print("Origin")
    case (0, y) :
        print(f"Y-axis at {y})")
    case (x, 0) :
        print(f"X-axis at {x})")
    case (x, y) :
        print(f"Point at ({x}, {y})")
    case _:    
        print("Not a valid point")
```

```python
# List length matching

items = ["apple", "banana", "cherry" ]
match items: 
    case []: 
        print("Empty list") 
    case [x]: 
        print(f"One item: {x}") 
    case [x, y]: 
        print(f"Two items: {x}, {y}") 
    case [x, y, *rest]: 
        print(f"First two: {x}, {y}, others: {rest}")
```

```python
# Dictionary matching (with case {"key":value})

data = {"name": "Alice", "age": 30, "city": "New York"}

match data:
    case {"name": name, "age": age}:
        print(f"Name: {name}, Age: {age}")
    case {"name": name}:
        print(f"Name: {name} Age not specified")
    case {"city": city}:
        print(f"City: {city}")
    case _:
        print("No match found")
```
